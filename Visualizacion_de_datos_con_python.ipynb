#**Visualizacion de datos con python: Matplotlib, Seaborn y Plotly**

```python
import pandas as pd
df = pd.read_csv('inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais',inplace=True)
df
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = df.loc['Colombia',anos]
colombia
```

```python
col_dict ={'Año':colombia.index.tolist(),
           'Inmigrantes':colombia.values.tolist()}
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'],datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.title('Inmigrantes de Colombia hacia Canada')
plt.xlabel('Año')
plt.ylabel('Numero de Imigrantes')
plt.show()
```

##**Crear figuras con Matplotlib**

```python
fix, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigracion de colombianos hacia Canada \n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Numero de Migrantes')
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))
axs[0].plot(datos_col['Año'],datos_col['Inmigrantes'])
axs[0].set_title('Imigracion de colombianos hacia Canada \n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de Migrantes')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de Migrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigracion de Colombianos \n hacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombianos')
axs[1].set_ylabel('Numero de Migrantes')
axs[1].grid()
plt.show()
```

##matplotlib

```python
fig, axs = plt.subplots(2,2,figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Imigracion de los 4 mayores paises sudamericanos\nhacia Canada desde 1980 hasta 2013 ')

axs[0,0].plot(df.loc['Colombia',anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil',anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú',anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)




for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Año')
  ax.set_ylabel('Numero de Migrantes')
  ax.grid()

plt.show()
```

```python
fix, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'],datos_col['Inmigrantes'],lw=2,color='g',marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigracion de colombianos hacia Canada \n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año',fontsize=14)
ax.set_ylabel('Numero de Migrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10= df.sort_values('Total',ascending=False).head(10)
top_10
```

```python
sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
plt.show()
```

```python
ax = sns.barplot(data=top_10,x='Total',y=top_10.index,orient='h')
ax.set(title='Top 10 países con mayor migración hacia Canadá \n1980 - 2013',
       xlabel='Numero de Imigrantes',
       ylabel='Año')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10,x='Total', y=top_10.index, orient='h')
ax.set_title('Top 10 países con mayor migración hacia Canadá \n1980 - 2013',
             loc='left',fontsize=16)
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
for i,j in enumerate(top_10['Total']):
  ax.text(j+20,i,str(j),color='black',fontsize=10,ha='left',va='center')
plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y',which='both',length=0)

  ax = sns.barplot(data=top_10,x='Total',y= top_10.index, hue=top_10.index, orient='h',
                   palette=palette,)
  ax.set_title('Top 10 países con mayor migración hacia Canadá \n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Imigrantes',fontsize=14)
  ax.set_ylabel('')
  plt.show()
```

```python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])



  ax = sns.barplot(data=top_10,x='Total',y= top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 países con mayor migración hacia Canadá \n1980 - 2013',
              loc='left',fontsize=16)
  ax.set_xlabel('Número de Imigrantes',fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black',fontsize=10,ha='left',va='center')
  plt.show()
```

```python
sns.set_theme()
generar_grafico('tab10')
```

```python
sns.set_theme(style='darkgrid')
generar_grafico('tab10')
```

```python
sns.set_theme(style='whitegrid')
generar_grafico('tab10')
```

```python
sns.set_theme(style='white')
generar_grafico('tab10')
```

```python
import plotly.express as px
```

```python
fig = px.line(datos_col,x='Año',y='Inmigrantes',
              title='Imigracion de colombianos hacia Canadá entre 1980 y 2013')
fig.update_layout(width=1000,height=500, xaxis={'tickangle':-45},
                  xaxis_title='Año', yaxis_title='Numero de Migrantes',
                  font_family='Arial', font_size=14, font_color='grey')
fig.show()
```

```python
fig = px.line(datos_col,x='Año',y='Inmigrantes',
              title='Imigracion de colombianos hacia Canadá entre 1980 y 2013')
fig.update_traces(line_color='green',line_width=3)
fig.update_layout(width=1000,height=500, xaxis={'tickangle':-45},
                  xaxis_title='Año', yaxis_title='Numero de Migrantes',
                  font_family='Arial', font_size=14, font_color='grey')
fig.show()
```
